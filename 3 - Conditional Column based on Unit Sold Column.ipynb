```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('csv').option('header', True).option('inferSchema', True).load('/FileStore/RawSource/sales_data_first.csv')
df.display()
```

```python
df = df.withColumn('Conditional Column', when(col('units_sold')==1, 'Low')\
                                        .when(col('units_sold')==2, 'Medium')\
                                        .otherwise('High'))
df.display()                                        
```

```python
def Conditional_Column(x):
    if (x==1):
        return 'Low'
    elif (x==2):
        return 'Medium'
    else :
        return 'High'
        
```

```python
Conditional_Column = udf(Conditional_Column)
```

```python
df = df.withColumn('Conditional Column New', Conditional_Column('units_sold'))
df.display()
```

```python
from pyspark.sql.window import Window
```

```python
df = df.withColumn('CumSum', sum('units_sold').over(Window.partitionBy('month').orderBy('month').rowsBetween(Window.unboundedPreceding, Window.currentRow)))
df.display()
```
